```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
import pytesseract
from pytesseract import Output
```

```python
def load_img():
    full = cv2.imread('D://10002.png')
    full = cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
    return full
```

```python
def display_img(full):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(full,cmap='gray')
```

```python
img = load_img()
kernel = np.ones(shape=(5,5),dtype=np.float32)/25
```

```python
kernel
```

```python
blurred = cv2.blur(img,ksize=(5,5))
display_img(blurred)
```

```python
ret,thresh1 = cv2.threshold(blurred,50,255,cv2.THRESH_TRUNC)
```
